```python
def z_score(df,field):
    data =df[field].to_numpy()
    mean=np.mean(data)
    std=np.std(data)
    return df.query(f"({field} - {mean})/{std}>3")
```

```python
field="dual_sim"
df=df.dropna(subset=[field])
df=df.sort_values(by=field)
documentation=z_score(df,field)
documentation
```
