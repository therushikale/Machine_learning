```python
import matplotlib.pyplot as plt
```

```python
x=[20,45,6,34,29,30,41,52,19,98,67,91,65,78]
y=[11,18,36,44,79,50,51,22,59,88,97,41,25,18]
```

```python
plt.plot(x,y,'s',ms=10)
plt.xlabel('Xdata')
plt.ylabel('Ydata')
```

```python
plt.plot(x,y,marker='+',ms=8)
```

```python
plt.plot(x,y,marker='+',ms=8,color='g')
```

```python
plt.plot(x,y,marker='+',ms=8,color='g',mfc='red',linestyle="dashed")
```

```python
plt.plot(x,y,marker='+',ms=8,color='g',mfc='red',linestyle="dashed",linewidth=3)
```

```python
plt.xlabel('Xdata',color='r')
plt.ylabel('Ydata',color='r')
plt.title('plot',loc='center',color='g')
plt.plot(x,y,marker='+',ms=8,color='g',mfc='red',linestyle="dashed",linewidth=3)
```
